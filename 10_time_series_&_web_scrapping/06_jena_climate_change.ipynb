# **Jena Climate Change Data Forecasting**

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
# load dataset
df = pd.read_csv("./data/jena_climate_2009_2016.csv")
df.head()
```

```python
# rename the columns to be used for prophet model
df = df.rename(columns={'Date Time': 'ds', 'T (degC)': 'y'})
df.head()
```

```python
from pandas import to_datetime

df['ds'] = to_datetime(df['ds'], format='%d.%m.%Y %H:%M:%S')
```

```python
df.head()
```

```python
# check the info of the data
df.info()
```

```python
# check the summary statistics of the data 
df.describe()
```

```python
# implement the prophet model
model = Prophet()

# fit the model
model.fit(df)

forecast = model.make_future_dataframe(periods=365)
forecast = model.predict(forecast)
plot_plotly(model, forecast)
```
