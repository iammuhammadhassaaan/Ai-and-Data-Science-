```python
# improt libraries
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px 
```

```python
# import titanic data
df = sns.load_dataset('titanic')
df.head()
```

# **1. Plots with Pandas**

```python
# make scatter plot with pandas 
df.plot.scatter(x = 'age', y = 'fare')
```

```python
# change color using hex codes
# make scatter plot with pandas 
df.plot.scatter(x='age', y='fare', color='#FF0000')
```

```python
# change color using hex codes
# make scatter plot with pandas 
df.plot(kind= 'scatter', x='age', y='fare', color='#FF0000')
```

```python
# make box plot with pandas with hex color
df.plot.box(color='#FF0000')
```

```python
# make plot in pandas with groupby function
df.groupby('sex').plot.scatter(x='age', y='fare', color='#FF0000')
```

```python
# make plot in pandas with groupby function
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

# **2. Seaborn Plots**

```python
#bar plot using seaborn on df
sns.barplot(x='sex', y='fare', data=df, hue='class')
```

```python
#bar plot using seaborn on df
sns.boxplot(x='sex', y='age', data=df, hue='class')
```

```python
#bar plot using seaborn on df
print(sns.scatterplot(x='age', y='fare', data=df, hue='class'))
```

```python
#bar plot using seaborn on df
sns.scatterplot(x='fare', y='age', data=df, hue='sex')
```

```python
# make scatter plot via plotly
import plotly.express as px

# Assuming you have a DataFrame called df with columns 'x' and 'y'
fig = px.scatter(df, x='age', y='fare', color='class')

fig.show()
```

```python
import plotly.express as px

# Load Gapminder dataset
df = px.data.gapminder()

# Create an animated scatter plot
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    animation_frame="year",
    log_x=True,  # Logarithmic scale for x-axis (gdpPercap)
    range_x=[300, 100000],  # Set x-axis range
    range_y=[25, 90],  # Set y-axis range
    title="Gapminder World Development",
    labels={"gdpPercap": "GDP per Capita", "lifeExp": "Life Expectancy"},
    size_max=60,  # Set maximum marker size
    hover_name="country",  # Show country name on hover
    facet_col="continent",  # Create subplots for each continent
    facet_col_wrap=3,  # Wrap subplots after 3 columns
)

# Show the animated plot
fig.show()
```
