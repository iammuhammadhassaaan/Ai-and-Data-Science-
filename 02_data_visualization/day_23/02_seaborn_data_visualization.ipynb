```python
# loading dataste
df = sns.load_dataset('titanic')
```

```python
df.info()
```

## 1. Scatter Plot

```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(16, 9))
sns.scatterplot(x='age', y='fare', data=df, hue='who')
```

```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(16, 9))
#change the color
who_palette = {'man': 'red', 'woman': 'green', 'child': 'blue'}
sns.scatterplot(x='age', y='fare', data=df, hue='who', palette=who_palette)
#labels
plt.xlabel('age')
plt.ylabel('fare')
#title
plt.title('Scatter Plot')

# saving a plot 150dpi, 300dpi, 600dpi
plt.savefig('scatter_plot.png', dpi=300)
```

```python
# style of the plot
sns.set_theme(style="darkgrid", font_scale = 1.5, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(8, 4))
#change the color
who_palette = {'man': 'red', 'woman': 'green', 'child': 'blue'}
sns.scatterplot(x='age', y='fare', data=df, hue='who', palette=who_palette)
#labels
plt.xlabel('age')
plt.ylabel('fare')
#title
plt.title('Scatter Plot')
```

Interpretation:
1. We have Age of passengers at x-axis and fare at y-axis
2. The scatter plot shows the relationship between age and fare of titanic passengers.
3. We further grouped the passenger by thier genders such as male(Green Points) and female (Red Points).
4. Two male passenger gave the highest fare in titanic and they might be travelling in First Class.
5. The oldest person at titanic is also a male having 80 years of age also shown in scatter plot. (As shown in Figure. 01)

```python
# style of the plot
sns.set_theme(style="darkgrid", font_scale = 1.5, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(16, 9))
#change the color
who_palette = {'man': 'red', 'woman': 'green', 'child': 'blue'}
sns.scatterplot(x='age', y='fare', data=df, hue='who', size='age' , palette=who_palette)
#labels
plt.xlabel('age')
plt.ylabel('fare')
#title
plt.title('Scatter Plot')
```

## 2. Boxplot

```python
df['class'].unique()
```

```python
# style of the plot
sns.set_theme(style="darkgrid", font_scale = 0.8, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(8, 4))
#change the color
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxplot(x='sex', y='age', data=df, hue='class', palette=palette)
#labels
plt.xlabel('Sex')
plt.ylabel('Age')
#title 
plt.title('Box Plot')
# change the position and location
plt.legend(loc='upper right', title = 'Class Name')
```

## 3. Boxen Plot

```python
# style of the plot
sns.set_theme(style="darkgrid", font_scale = 2 , font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(20, 9))
#change the color
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(x='sex', y='age', data=df, hue='class', palette=palette)
#labels
plt.xlabel('Sex')
plt.ylabel('Age')
#title 
plt.title('Boxen Plot')
# change the position and location
plt.legend(loc='upper right', title = 'Class Name')
# change the xtick labels
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation = 0)
```

```python
# make scatter plot with plotly 
import plotly.express as px 

fig = px.scatter(df, x='age', y='fare', color = 'sex')
fig.show()
```

## 4. Barplot

```python
# style of the plot
sns.set_theme(style="darkgrid", font_scale = 2 , font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(20, 9))
#change the color
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.barplot(x='sex', y='age', data=df, hue='class', palette=palette, capsize=0.1)
#labels
plt.xlabel('Sex')
plt.ylabel('Age')
#title 
plt.title('Bar Plot')
# change the position and location
plt.legend(loc='upper right', title = 'Class Name')
# change the xtick labels
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation = 0)
```

```python
# style of the plot
sns.set_theme(style="darkgrid", font_scale = 2 , font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(20, 9))
#change the color
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.violinplot(x='sex', y='age', data=df, hue='class', palette=palette)
#labels
plt.xlabel('Sex')
plt.ylabel('Age')
#title 
plt.title('Bar Plot')
# change the position and location
plt.legend(loc='upper right', title = 'Class Name')
# change the xtick labels
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation = 0)
```
