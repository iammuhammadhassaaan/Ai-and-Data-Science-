# **Heirarchichal Clustering**
* Hierarchical clustering is a type of unsupervised machine learning algorithm that builds a hierarchy of clusters by merging or splitting existing clusters based on their similarity. It can be used for both agglomerative and divisive clustering.

```python
# improt libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# generate a random dataset with 100 samples and 3 features 
X, y = make_blobs(n_samples=100, centers=50, n_features=3, random_state=42)

# perform heirarchichal clustering with the Agglomerative Clustering
# Define the clustering model with clear parameter names
clustering = AgglomerativeClustering(
    n_clusters=3,  # Number of clusters to form
    linkage='ward'  # Linkage criterion to use
)

# fit the model to the data
clustering.fit(X)

# plot he dendogram 
plt.figure(figsize=(10, 8))
plt.title('Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```

```python
# apply on tips dataset 
df = sns.load_dataset('tips')

# Select the relevant features for clustering
features = df[['total_bill', 'tip', 'size']]

# Create a sample dataset with 10 clusters using make_blobs
X, _ = make_blobs(n_samples=len(features), centers=50, n_features=3, random_state=42)

# Define the clustering model with clear parameter names
clustering = AgglomerativeClustering(
    n_clusters=5,  # Number of clusters to form
    linkage='ward'  # Linkage criterion to use
)

# fit the model to the data
clustering.fit(X)

# plot he dendogram 
plt.figure(figsize=(10, 8))
plt.title('Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```

```python
# apply on tips dataset 
df = sns.load_dataset('tips')

# Select the relevant features for clustering
features = ['total_bill', 'tip', 'size']

# Create a sample dataset with 10 clusters using make_blobs
X = df[features]

# compute the linkage matrix
linkage_matrix = linkage(X, method='ward')

# plot he dendogram 
plt.figure(figsize=(10, 8))
sch.dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(X, cmap='coolwarm', annot=True)
plt.title('Heatmap')
plt.xlabel('Features')
plt.ylabel('Samples')
plt.show()
```

```python
# apply on tips dataset 
df = sns.load_dataset('tips')

# Select the relevant features for clustering
features = ['total_bill', 'tip', 'size']

# Create a sample dataset with 10 clusters using make_blobs
X = df[features]

# transpose the data matrix
X_transposed = X.transpose()

# compute the linkage matrix
linkage_matrix = linkage(X_transposed, method='ward')

# plot he dendogram 
plt.figure(figsize=(10, 8))
sch.dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```
