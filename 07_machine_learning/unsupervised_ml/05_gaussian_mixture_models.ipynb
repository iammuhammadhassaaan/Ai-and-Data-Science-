# Gaussian Mixture Models (GMMs) Notebook

## Introduction

Gaussian Mixture Models (GMMs) are probabilistic models that assume all data points are generated from a mixture of several Gaussian distributions. They are widely used for clustering, density estimation, and anomaly detection.

### Key Concepts

1. **Gaussian Distribution**: A bell-shaped curve defined by its mean and variance.
2. **Mixture Model**: A combination of several Gaussian distributions.
3. **Expectation-Maximization (EM)**: An algorithm used to estimate the parameters of a GMM.

## Mathematical Foundation

### 1. Gaussian Distribution

The Gaussian distribution (or normal distribution) is defined as:

$$ \mathcal{N}(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( -\frac{(x - \mu)^2}{2 \sigma^2} \right) $$

where:
\begin{align*}
    \mu &\text{ is the mean}, \\
    \sigma^2 &\text{ is the variance}.
\end{align*}


### 2. Mixture of Gaussians

A Gaussian Mixture Model assumes the data is generated from a mixture of \( k \) Gaussian distributions:

$$ p(x) = \sum_{i=1}^k \pi_i \mathcal{N}(x | \mu_i, \sigma_i^2) $$

where:
\begin{align*}
    \pi_i &\text{ is the weight of the } i\text{-th Gaussian}, \\
    \mu_i &\text{ and } \sigma_i^2 \text{ are the mean and variance of the } i\text{-th Gaussian}.
\end{align*}



## Expectation-Maximization (EM) Algorithm

The EM algorithm is used to find the maximum likelihood estimates of the parameters in a GMM.

### 1. Expectation Step (E-Step)

Calculate the probability that each data point belongs to each Gaussian component.

### 2. Maximization Step (M-Step)

Update the parameters (weights, means, variances) based on the probabilities computed in the E-Step.

### 3. Convergence

Repeat the E-Step and M-Step until the parameters converge (i.e., the changes are smaller than a threshold).


## Applications

1. **Clustering**: Grouping data into clusters.
2. **Density Estimation**: Estimating the probability density function of the data.
3. **Anomaly Detection**: Identifying unusual data points.

```python
# import libraies 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.mixture import GaussianMixture

# load iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]

# create a GMM model with clusters
gmm = GaussianMixture(n_components=3)

# fit the model
gmm.fit(X)

# predict the labels
labels = gmm.predict(X)

# visualize the clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
# import libraies 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.mixture import GaussianMixture

# load iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]

# create a GMM model with clusters
gmm = GaussianMixture(n_components=3)

# fit the model
gmm.fit(X)

# predict the labels
labels = gmm.predict(X)

# compute the log-likelihood
log_likelihood = gmm.score(X)

# print the log-likelihood
print("Log-likelihood:", log_likelihood)

# visualize the clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
# import libraies 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.mixture import GaussianMixture

# load iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]

# create a GMM model with clusters
gmm = GaussianMixture(n_components=3)

# fit the model
gmm.fit(X)

# predict the labels
labels = gmm.predict(X)

# compute the log-likelihood
log_likelihood = gmm.score(X)

# compute the AIC and BIC
n_samples, _ = X.shape
aic = gmm.aic(X) # akaike information criterion
bic = gmm.bic(X) # bayesian information criterion

# print the evaluation matrix
print("Log-likelihood:", log_likelihood)
print("AIC:", aic)
print("BIC:", bic)


# visualize the clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```
