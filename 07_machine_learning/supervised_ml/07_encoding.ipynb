# **Feature Encoding**

```python
# import libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# load the data 
df = sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

```python
df['smoker'].value_counts()
```

```python
# lets encode the time in label encoder with sklearn 
# import libraies 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder 

le = LabelEncoder()

df['encoded_time'] = le.fit_transform(df['time'])

df.head()
```

```python
# ordinal encoding 

oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])

df['encoded_day'] = oe.fit_transform(df[['day']])

df.sample(10)
```

```python
# one hot encoding 
ohe = OneHotEncoder()
ohe.fit_transform(df[['day']]).toarray()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

# Load the titanic dataset
df = sns.load_dataset('titanic')

# One-hot encode the 'embarked' column
ohe = OneHotEncoder()
embarked_ohe = pd.DataFrame(ohe.fit_transform(df[['embarked']]).toarray(), 
                             columns=ohe.get_feature_names_out(['embarked']))

# Concatenate the original dataframe with the one-hot encoded dataframe
df = pd.concat([df, embarked_ohe], axis=1)

df.head()
```
