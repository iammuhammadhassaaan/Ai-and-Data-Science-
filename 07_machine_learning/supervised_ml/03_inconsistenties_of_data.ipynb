```python
import pandas as pd
import numpy as np

# Create a dictionary with inconsistent data
data = {
    'date': ['2021-12-01', '1-12-2024', '2022/12/01', '12-01-2022'],
    'country': ['USA', 'United States', 'America', 'U.S.A.'],
    'name': ['John', 'john', 'jonh', 'jane'],
    'sales': [100, 200, 400,'300']
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

df.head()
```

```python
# standardizing the dateformat
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
# harmonize the name of the data 
country_mapping = {'USA': 'United States', 'U.S.A.': 'United States', 'America': 'United States'}
df['country'] = df['country'].replace(country_mapping)
df.head()
```

```python
# correct the typographical examples in name 
df['name'] = df['name'].replace({'john': 'John', 'jane': 'John'})
df.head()
```
