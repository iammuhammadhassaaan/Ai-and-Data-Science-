```python
# How much rural population increased from 1998 to 2017?
pop_rural_1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural_2017 = df['ALL SEXES (RURAL)'].sum()

# How much increased?
increase_pop_rural = pop_rural_2017 - pop_rural_1998
print("The increase in the rural population from 1998 to 2017: ", increase_pop_rural)
```

```python
# How much rural population increased from 1998 to 2017?
pop_urban_1998 = df['POPULATION 1998 (URBAN)'].sum()
pop_urban_2017 = df['ALL SEXES (URBAN)'].sum()

# How much increased?
increase_pop_urban = pop_urban_2017 - pop_urban_1998
print("The increase in the urban population from 1998 to 2017: ", increase_pop_urban)
```
