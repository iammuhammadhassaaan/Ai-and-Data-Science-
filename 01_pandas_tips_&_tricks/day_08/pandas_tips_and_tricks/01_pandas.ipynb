## **03- HOw to rename columns?**

```python
# create a list of dictionaries
data = [{'Name': 'Tom', 'Age': 27, 'City': 'New York'},
        {'Name': 'Nick', 'Age': 31, 'City': 'London'},
        {'Name': 'John', 'Age': 19, 'City': 'Paris'},
        {'Name': 'Julia', 'Age': 32, 'City': 'Madrid'}]

# create a dataframe
df = pd.DataFrame(data)

# display the dataframe
print(df)
```

```python
# Rename the columns
df = df.rename(columns={'Name': 'Person', 'Age': 'Umar', 'City': 'Location'})

# display the dataframe
df
```

```python
df.columns=['Banda', 'Umar', 'Jagha']
df
```

```python
df.info()
```

```python
# to replace any charachter, string

df.columns=df.columns.str.replace('Banda', 'banda')
df.columns=df.columns.str.replace('Umar', 'umar')

df
```

```python
# Adding prefix to columns
df = df.add_prefix('baba_')
print(df)
```

```python
# adding suffix to a columns
df = df.add_suffix("_baba")
df
```

## **16- Spliting a string into a multiple columns**

```python
import pandas as pd

df = pd.DataFrame({'name': ['Ahmad Raza', 'Ali Afzal', 'Sajid Ahmad'],
        'location': ['Multan, Pakistan', 'Lahore, Pakistan', 'Islamabad, Pakistan']})

df
```

```python
df.name.str.split(' ', expand=True)
```

```python
df[["first_name", "last_name"]] = df.name.str.split(' ', expand=True)
```

```python
df
```

```python
df[["city_name", "country_name"]] = df.location.str.split(',', expand=True)
```

```python
df
```

```python
df = df[['first_name', 'last_name', 'city_name', 'country_name']]
df
```
