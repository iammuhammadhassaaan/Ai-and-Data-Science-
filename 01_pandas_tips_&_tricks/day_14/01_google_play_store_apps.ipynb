# Check out that how many MBs are there n one kb? and how to do that 

```python
# Function to convert KB to MB
def convert_to_MB(value):
    if 'k' in value:
        return float(value.replace('k', '')) / 1024
    elif 'M' in value:
        return float(value.replace('M', ''))

# Applying the conversion function to the 'size' column
df['Size'] = df['Size'].apply(convert_to_MB)

print(df)
```

## 2. Install

# remove the plus sign and create binning 

```python
# Function to clean installs and convert to numeric
def clean_installs(value):
    return int(value.replace(',', '').replace('+', ''))

# Apply function to 'Installs' column
df['Installs'] = df['Installs'].apply(clean_installs)
# Print the first 5 rows of the DataFrame
print(df.head())
```

```python
df.head()
```

```python
# Define bins
bins = [0, 1000, 10000, 500000, 1000000, df['Installs'].max()]

# Define labels for the bins
labels = ['0-1k', '1k-10k', '10k-500k', '500k-1M', '1M+']

# Create a new column 'Installs Group' with bins
df['Installs Group'] = pd.cut(df['Installs'], bins=bins, labels=labels)

df.head()
```
