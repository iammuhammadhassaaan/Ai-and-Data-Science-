# Distributions
---

```python
import numpy as np 
import matplotlib.pyplot as plt

# Generating example datasets with different distributions
np.random.seed(0) # For reproducibility

# Normal distribution data
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Uniform distribution data
uniform_data = np.random.uniform(low=-2, high=2, size=1000)

# Exponential distributiondata
exponential_data = np.random.exponential(scale=1, size=1000)

# plotting the distributions
plt.figure(figsize=(12,8))

# Normal Distribution PLot
plt.subplot(3, 1, 1)
plt.hist(normal_data, bins=30, color='skyblue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')

# Uniform Distribution Plot
plt.subplot(3,  1, 2)
plt.hist(uniform_data, bins=30, color='lightgreen', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')

# Exponential Distribution Plot
plt.subplot(3,1,3)
plt.hist(exponential_data, bins=30, color='salmon', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Exponential Distribution')

plt.tight_layout()
plt.show()
```

# **Skewness and Kurtosis**
---

```python
df_titanic.head()
```

```python
# find kurtosis of age
stats.kurtosis(df_titanic['age'], fisher=False)
```

```python
# find kurtosis of fare 
stats.kurtosis(df_titanic['fare'], fisher=False)
```

```python
sns.histplot(df_titanic, x = 'fare', kde=True)
```

```python
# find kurtosis of fare
stats.skew(df_titanic['fare'])
```

```python
import numpy as np 
from scipy.stats import skew, kurtosis

# Generating example data
np.random.seed(0)
data = np.random.normal(0, 1, 1000) # Normal Distribution

# Calculating skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data, fisher=True) # fisher=True gives the acess kurtosis

# Displaying the result
print(f'Skewness of the dataset: {data_skewness}')
print(f'Kurtosis of the dataset: {data_kurtosis}')
```

```python
sns.histplot(data=data, kde=True)
```

```python
df_third = df_titanic[df_titanic['class'] == 'Third']
stats.kurtosis(df_third['fare'], fisher=False)  # Kurtosis of the fare column
```

```python
stats.skew(df_third['fare'])
```

```python
sns.histplot(data=df_third, x = 'fare', kde=True)
```

```python
# how to test the normal distribution of the data

# H0: data is normally distributed
# H1: data is not normally distributed

# shpiro wilk
# D'Agostino's K^2 Test
# Anderson-Darling Test
```

```python
# shapiro wilk test to test the normality of the data
stats.shapiro(data)
print(f"Shapiro-Wilk test statistic: {stats.shapiro(data)[0]}")
print(f"Shapiro-Wilk test p-value: {stats.shapiro(data)[1]}")
```
